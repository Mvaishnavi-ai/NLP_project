<h1>Fog index</h1>

```python
def average_sentence_length(tokens, num_sentences):
    total_words = len(tokens)
    return total_words / num_sentences if num_sentences > 0 else 0

def estimate_sentences(tokens, avg_sentence_length=15):
    return max(1, len(tokens) // avg_sentence_length)

def count_syllables(word):
    word = word.lower()
    if word.endswith('es') or word.endswith('ed'):  # Remove 'es', 'ed' at the end of the word
        word = word[:-2]
    vowels = "aeiou"  # Count vowels
    syllable_count = len([char for char in word if char in vowels])
    return max(1, syllable_count)

def percentage_complex_words(tokens):
    complex_word_count = sum(1 for word in tokens if count_syllables(word) > 2)  # Count complex words (syllable > 2)
    return (complex_word_count / len(tokens)) * 100 if len(tokens) > 0 else 0 

def calculate_fog_index(tokens, num_sentences):
    asl = average_sentence_length(tokens, num_sentences)
    pcw = percentage_complex_words(tokens)
    return 0.4 * (asl + pcw)

num_sentences_per_row = [estimate_sentences(row, avg_sentence_length=15) for row in tokens]

fog_indices = [calculate_fog_index(row, num_sentences) for row, num_sentences in zip(tokens, num_sentences_per_row)]
for idx, fog_index in enumerate(fog_indices):
    print(f"Row {idx + 1}: Fog Index = {fog_index:.2f}")

```

```python
import string 
stop_words = set(stopwords.words('english')) # Load NLTK stopwords

def cleaned_word_count(tokens):
    # Remove punctuation
    tokens = [word.translate(str.maketrans('', '', string.punctuation)) for word in tokens]
    # Remove stopwords
    tokens = [word for word in tokens if word.lower() not in stop_words]
    # Count remaining words
    return len([word for word in tokens if word.strip() != ''])  # Ensure no empty strings

# Calculate the word count for each row
word_counts = tokens.apply(cleaned_word_count)

for idx, count in enumerate(word_counts):
    print(f"Row {idx + 1}: Word Count = {count}")

```

<h1> Personal Pronouns</h1>

```python
personal_pronouns = ["I", "we", "my", "ours", "us"]

def count_personal_pronouns(tokens):
     text = " ".join(tokens).lower() # Convert to lowercase
     text = re.sub(r'\bus\b', '', text)
     count = 0
     for pronoun in personal_pronouns:
        count += len(re.findall(r'\b' + re.escape(pronoun) + r'\b', text))
     return count

# Calculate the personal pronoun count 
personal_pronoun_counts = [count_personal_pronouns(row) for row in tokens]

for idx, count in enumerate(personal_pronoun_counts):
    print(f"Row {idx + 1}: Personal Pronouns Count = {count}")
```
